# Pandas Data Analysis Case Study: Titanic & Tips Datasets
Bu çalışma, veri analitiği süreçlerinde sıkça kullanılan **Pandas** kütüphanesi ile veri manipülasyonu, temizleme ve analiz tekniklerini içermektedir.

### İçerik:
1. **Car Crashes:** List Comprehension ile sütun ismi düzenleme.
2. **Titanic:** Eksik veri yönetimi ve demografik analiz.
3. **Tips:** Gruplama (Groupby) ve gelişmiş filtreleme (loc) teknikleri.

import seaborn as sns
import pandas as pd
import numpy as np

# Veri setlerini yükleyelim
df_car = sns.load_dataset("car_crashes")
df_titanic = sns.load_dataset("titanic")
df_tips = sns.load_dataset("tips")

## 1. Car Crashes Veri Seti - Sütun Manipülasyonu
Bu bölümde List Comprehension kullanarak sütun isimlerini belirli kurallara göre standartlaştırıyoruz.

# Sayısal sütunların başına "NUM_" ekleyip tüm sütun isimlerini büyütmek
df_car.columns = ["NUM_" + col.upper() if df_car[col].dtypes != "O" else col.upper() for col in df_car.columns]

# İçinde "NO" geçmeyen sütunların sonuna "_FLAG" eklemek
df_car.columns = [col.upper() + "_FLAG" if "NO" not in col.upper() else col.upper() for col in df_car.columns]

df_car.head()

## 2. Titanic Veri Seti - Veri Temizleme ve Filtreleme
Yolcu verileri üzerinde eksik değer analizi yapıyor ve stratejik filtreleme yöntemlerini uyguluyoruz.

# Embarked sütun tipini optimize etme
df_titanic["embarked"] = df_titanic["embarked"].astype("category")

# Eksik Değerlerin Mod ve Medyan ile Doldurulması
df_titanic["deck"].fillna(df_titanic["deck"].mode()[0], inplace=True)
df_titanic["age"].fillna(df_titanic["age"].median(), inplace=True)

# Filtreleme Örneği: 30 Yaş altı kadın yolcular
young_female_df = df_titanic[(df_titanic["sex"] == "female") & (df_titanic["age"] < 30)]

# Gruplama: Sınıf ve Cinsiyete göre hayatta kalma oranları
survival_stats = df_titanic.groupby(["pclass", "sex"]).agg({"survived": ["sum", "count", "mean"]})
survival_stats

## 3. Tips Veri Seti - Finansal Analiz ve Gruplama
Bahşiş verileri üzerinden müşteri harcama alışkanlıklarını ve servis zamanlarını analiz ediyoruz.

# Gün ve Zaman bazlı toplam hesap ve bahşiş istatistikleri
time_stats = df_tips.groupby(["time", "day"]).agg({
    "total_bill": ["sum", "min", "max", "mean"],
    "tip": ["sum", "min", "max", "mean"]
})

# loc ile Filtreleme: Masa boyutu 3'ten küçük ve hesap 10'dan büyük olanlar
mean_bill = df_tips.loc[(df_tips["size"] < 3) & (df_tips["total_bill"] > 10), "total_bill"].mean()

# Yeni Değişken Oluşturma ve Sıralama
